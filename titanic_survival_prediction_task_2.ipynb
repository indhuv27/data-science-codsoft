IMPORTING THE DEPENDENCIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

DATA PROCESSING

```python
# load the data into pandas dataset
titanic_data = pd.read_csv('Titanic-Dataset.csv')
```

```python
# printing the first 5 rows
titanic_data.head()
```

```python
# number of rows and columns
titanic_data.shape
```

```python
# getting some information about the data
titanic_data.info()
```

```python
# check the number of missing values in each columns
titanic_data.isnull().sum()
```

HANDLING THE MISSING VALUES

```python
# drop the 'Cabin' column from the dataframe
titanic_data = titanic_data.drop(columns='Cabin', axis=1)
```

```python
# replacing the missing values in 'age' column with mean values
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

```python
# finding the mode value of 'embarked'column
print(titanic_data['Embarked'].mode())
```

```python
print (titanic_data['Embarked'].mode()[0])
```

```python
# replacing the missing values in each column
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
# check the number of missing values
titanic_data.isnull().sum()
```

DATA ANALYSIS

```python
# getting some statistical measures about the data
titanic_data.describe()
```

```python
# finding the number of pepoles survived and not survived
titanic_data['Survived'].value_counts()
```

DATA VISUALISATION

```python
# making a counter plot for 'survival column'
sns.countplot(x='Survived', data=titanic_data)
plt.show()
```

```python
# number of survivors gender wise
sns.countplot(x='Sex', hue='Survived', data=titanic_data)
plt.show()
```

```python
# number of count plot for 'Pclass' column
sns.countplot(x='Pclass', data=titanic_data)
plt.show()
```

ENCODING THE CATEGORICAL COLUMN


```python
titanic_data['Sex'].value_counts()
```

```python
titanic_data['Embarked'].value_counts()
```

```python
# converting the categorical column
titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_data.head()
```

SEPARATING FEATURES AND TARGETS

```python
X = titanic_data.drop(columns = ['PassengerId','Name','Ticket','Survived'],axis=1)
Y = titanic_data['Survived']
```

```python
print(X)
```

```python
print(Y)
```

SPLITTING THE DATA INTO TRAIN DATA AND TEST DATA

```python

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MODEL TRAINING

```python
model = LogisticRegression()
```

```python
# training the logistic Regression model with training data
model.fit(X_train, Y_train)
```

MODEL EVALUATION

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)

```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
